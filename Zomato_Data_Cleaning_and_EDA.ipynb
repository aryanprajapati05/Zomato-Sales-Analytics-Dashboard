# Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Zomato_Data.xlsx')
```

```python
df['Order_Amount']
```

```python
df['Order_Amount'] = pd.to_numeric(df['Order_Amount'], errors='coerce')
```

```python
df['Order_Amount']=df['Order_Amount'].fillna(df['Order_Amount'].mean())
```

```python
df.isnull().sum()
```

```python
df['Order_Date']=pd.to_datetime(df['Order_Date'])
```

```python
df.info()
```

```python
df['Customer_Age']=pd.to_numeric(df['Customer_Age'],errors='coerce')
```

```python
df['Customer_Age']=df['Customer_Age'].fillna(df['Customer_Age'].median())
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.dropna(subset=['Order_ID'])
```

```python
df['City']=df['City'].fillna('NA')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Delivery_Time_Min']=df['Delivery_Time_Min'].fillna(df['Delivery_Time_Min'].mean())
```

```python
df['Distance_km']=df['Distance_km'].fillna(df['Delivery_Time_Min'].mean())
```

```python
df['Discount_Percent'] = df['Discount_Percent'].fillna(0)
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].median())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df['Customer_Name']=df['Customer_Name'].str.strip().str.title()
```

```python
(df['Customer_Name'] == '').sum()
```

```python
df['Customer_Name'] = df['Customer_Name'].replace('', np.nan)
```

```python
df.info()
```

```python
df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')
```

```python
df['City']=df['City'].str.strip().str.title()
```

```python
df['Order_ID'].str.strip()
```

```python
df[df['Order_ID'] == '']
```

```python
df['Restaurant_Name']=df['Restaurant_Name'].str.strip().str.title()
```

```python
df['Cuisine']=df['Cuisine'].str.strip().str.title()
```

```python
df['Payment_Method']=df['Payment_Method'].str.strip().str.title()
```

```python
df['Delivery_Status']=df['Delivery_Status'].str.strip().str.title()
```

Performed comprehensive data cleaning by handling missing values, converting data types, standardizing categorical fields, removing hidden blank values, validating data quality, and engineering date-based features for analysis.

```python
df.to_csv('Zomato_Cleaned_Data.csv', index=False)
```
